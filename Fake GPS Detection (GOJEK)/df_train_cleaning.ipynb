The clean data will provide informations that extracted from original dataset. The extracted informations were decided by user based on consideration if the extracted information will give good correllation. The extraxted information (features) are:

label                fake or not
accuracy_max         the highest difference in accuracy at each order
accuracy_min         the lowest difference in accuracy at each order
accuracy_avg         the average of difference in accuracy at each order
accuracy_range       last accuracy - first accuracy at each order
longitude_max        the highest difference in longitude at each order
longitude_min        the lowest difference in longitude at each order
longitude_avg        the average of difference in longitude at each order
longitude_range      last longitude - first longitude at each order
latitude_max         the highest difference in latitude at each order
latitude_min         the lowest difference in latitude at each order
latitude_avg         the average of difference in latitude at each order
latitude_range       last latitude - first latitude at each order
altitude_max         the highest difference in altitude at each order
altitude_min         the lowest difference in altitude at each order
altitude_avg         the average of difference in altitude at each order
altitude_isnan       how many altitude=nan provide at each order
seconds_max          the highest difference in seconds at each order
seconds_min          the lowest difference in seconds at each order
seconds_avg          the average of difference in seconds at each order
seconds_range        last seconds - first seconds at each order
date_nunique         number of distict date at each order
hour_nunique         number of distict hour at each order
AVAILABLE            number of driver status is AVAILABLE at each order
OTW_DROPOFF          number of driver status is OTW_DROPOFF at each order
OTW_PICKUP           number of driver status is OTW_PICKUP at each order
UNAVAILABLE          number of driver status is UNAVAILABLE at each order
GO_FOOD              GO_FOOD service
GO_RIDE              GO_RIDE service
pings                number of pings at each order

```python
label=df[['order_id','label']]
label=label.groupby('order_id').label.max().reset_index()
label
```

```python
accuracy=df[['order_id','accuracy_in_meters']]
accuracy
```

```python
accuracy_max=accuracy.groupby('order_id').accuracy_in_meters.max().reset_index()
accuracy_max= accuracy_max.rename(columns={'accuracy_in_meters': 'accuracy_max'}) 
accuracy_max

accuracy_min=accuracy.groupby('order_id').accuracy_in_meters.min().reset_index()
accuracy_min= accuracy_min.rename(columns={'accuracy_in_meters': 'accuracy_min'})
accuracy_min

accuracy_avg=accuracy.groupby('order_id').accuracy_in_meters.mean().reset_index()
accuracy_avg= accuracy_avg.rename(columns={'accuracy_in_meters': 'accuracy_avg'})
accuracy_avg

accuracy_all=pd.merge(accuracy_max,accuracy_min,on=['order_id','order_id'])
accuracy_all=pd.merge(accuracy_all,accuracy_avg,on=['order_id','order_id'])
accuracy_all['accuracy_range']=accuracy_all['accuracy_max']-accuracy_all['accuracy_min']
accuracy_all
```

```python
longitude=df[['order_id','longitude']]
longitude
```

```python
longitude['interval']=longitude.groupby('order_id')['longitude'].diff()
pd.options.mode.chained_assignment = None
longitude
```

```python
longitude_max=longitude.groupby('order_id').interval.max().reset_index()
longitude_max=longitude_max.rename(columns={'interval': 'longitude_max'}) 
longitude_max

longitude_min=longitude.groupby('order_id').interval.min().reset_index()
longitude_min=longitude_min.rename(columns={'interval': 'longitude_min'}) 
longitude_min

longitude_avg=longitude.groupby('order_id').interval.mean().reset_index()
longitude_avg=longitude_avg.rename(columns={'interval': 'longitude_avg'}) 
longitude_avg

longitude_range_head=longitude.drop('interval',axis=1)
longitude_range_head=longitude_range_head.groupby('order_id').head(1)
longitude_range_tail=longitude.drop('interval',axis=1)
longitude_range_tail=longitude_range_tail.groupby('order_id').tail(1)
longitude_range=pd.merge(longitude_range_head,longitude_range_tail,on=['order_id','order_id'])
longitude_range['longitude_range']=longitude_range['longitude_y']-longitude_range['longitude_x']
longitude_range=longitude_range.drop(['longitude_y','longitude_x'],axis=1)
longitude_range


longitude_all=pd.merge(longitude_max,longitude_min,on=['order_id','order_id'])
longitude_all=pd.merge(longitude_all,longitude_avg,on=['order_id','order_id'])
longitude_all=pd.merge(longitude_all,longitude_range,on=['order_id','order_id'])
longitude_all
```

```python
latitude=df[['order_id','latitude']]
latitude
```

```python
latitude['interval']=latitude.groupby('order_id')['latitude'].diff()
pd.options.mode.chained_assignment = None
latitude
```

```python
latitude_max=latitude.groupby('order_id').interval.max().reset_index()
latitude_max=latitude_max.rename(columns={'interval': 'latitude_max'}) 
latitude_max

latitude_min=latitude.groupby('order_id').interval.min().reset_index()
latitude_min=latitude_min.rename(columns={'interval': 'latitude_min'}) 
latitude_min

latitude_avg=latitude.groupby('order_id').interval.mean().reset_index()
latitude_avg=latitude_avg.rename(columns={'interval': 'latitude_avg'}) 
latitude_avg

latitude_range_head=latitude.drop('interval',axis=1)
latitude_range_head=latitude_range_head.groupby('order_id').head(1)
latitude_range_tail=latitude.drop('interval',axis=1)
latitude_range_tail=latitude_range_tail.groupby('order_id').tail(1)
latitude_range=pd.merge(latitude_range_head,latitude_range_tail,on=['order_id','order_id'])
latitude_range['latitude_range']=latitude_range['latitude_y']-latitude_range['latitude_x']
latitude_range=latitude_range.drop(['latitude_y','latitude_x'],axis=1)
latitude_range


latitude_all=pd.merge(latitude_max,latitude_min,on=['order_id','order_id'])
latitude_all=pd.merge(latitude_all,latitude_avg,on=['order_id','order_id'])
latitude_all=pd.merge(latitude_all,latitude_range,on=['order_id','order_id'])
latitude_all
```

```python
altitude=df[['order_id','altitude_in_meters']]
altitude
```

```python
altitude['interval']=altitude.groupby('order_id')['altitude_in_meters'].diff()
pd.options.mode.chained_assignment = None
altitude
```

```python
altitude_max=altitude.groupby('order_id').interval.max().reset_index()
altitude_max=altitude_max.rename(columns={'interval': 'altitude_max'}) 
altitude_max

altitude_min=altitude.groupby('order_id').interval.min().reset_index()
altitude_min=altitude_min.rename(columns={'interval': 'altitude_min'}) 
altitude_min

altitude_avg=altitude.groupby('order_id').interval.mean().reset_index()
altitude_avg=altitude_avg.rename(columns={'interval': 'altitude_avg'}) 
altitude_avg


altitude_isnan=altitude[['order_id','altitude_in_meters']]
altitude_isnan['altitude_isnan']=altitude_isnan.altitude_in_meters.isnull()
altitude_isnan=altitude_isnan.groupby('order_id')['altitude_isnan'].sum().reset_index()
altitude_isnan


altitude_all=pd.merge(altitude_max,altitude_min,on=['order_id','order_id'])
altitude_all=pd.merge(altitude_all,altitude_avg,on=['order_id','order_id'])
altitude_all=pd.merge(altitude_all,altitude_isnan,on=['order_id','order_id'])
pd.set_option('display.max_rows', 5000)
altitude_all.head()
```

```python
seconds=df[['order_id','seconds']]
seconds
```

```python
seconds['interval']=seconds.groupby('order_id')['seconds'].diff()
pd.options.mode.chained_assignment = None
seconds
```

```python
seconds_max=seconds.groupby('order_id').interval.max().reset_index()
seconds_max=seconds_max.rename(columns={'interval': 'seconds_max'}) 
seconds_max

seconds_min=seconds.groupby('order_id').interval.min().reset_index()
seconds_min=seconds_min.rename(columns={'interval': 'seconds_min'}) 
seconds_min

seconds_avg=seconds.groupby('order_id').interval.mean().reset_index()
seconds_avg=seconds_avg.rename(columns={'interval': 'seconds_avg'}) 
seconds_avg

seconds_range_head=seconds.drop('interval',axis=1)
seconds_range_head=seconds_range_head.groupby('order_id').head(1)
seconds_range_tail=seconds.drop('interval',axis=1)
seconds_range_tail=seconds_range_tail.groupby('order_id').tail(1)
seconds_range=pd.merge(seconds_range_head,seconds_range_tail,on=['order_id','order_id'])
seconds_range['seconds_range']=seconds_range['seconds_y']-seconds_range['seconds_x']
seconds_range=seconds_range.drop(['seconds_y','seconds_x'],axis=1)
seconds_range


seconds_all=pd.merge(seconds_max,seconds_min,on=['order_id','order_id'])
seconds_all=pd.merge(seconds_all,seconds_avg,on=['order_id','order_id'])
seconds_all=pd.merge(seconds_all,seconds_range,on=['order_id','order_id'])
seconds_all.head()
```

```python
status=df[['order_id','driver_status']]
status
```

```python
dummies= pd.get_dummies(status['driver_status'],drop_first=False)  
status=pd.concat([status.drop('driver_status', axis=1), dummies],axis=1) 
status
```

```python
status=status.groupby('order_id')[['AVAILABLE','OTW_DROPOFF','OTW_PICKUP','UNAVAILABLE']].sum().reset_index()
status.head()
```

```python
service=df[['order_id','service_type']]
service
```

```python
dummies= pd.get_dummies(service['service_type'],drop_first=False)  
service=pd.concat([service.drop('service_type', axis=1), dummies],axis=1) 
service
```

```python
service=service.groupby('order_id')[['GO_FOOD','GO_RIDE']].max().reset_index()
service.head()
```

```python
pings=df[['order_id']]
pings['pings']=1
pings
```

```python
pings=pings.groupby('order_id')['pings'].sum().reset_index()
pings.head()
```

```python
clean_df=pd.merge(label,accuracy_all,on=['order_id','order_id'])
clean_df=pd.merge(clean_df,longitude_all,on=['order_id','order_id'])
clean_df=pd.merge(clean_df,latitude_all,on=['order_id','order_id'])
clean_df=pd.merge(clean_df,altitude_all,on=['order_id','order_id'])
clean_df=pd.merge(clean_df,seconds_all,on=['order_id','order_id'])
clean_df=pd.merge(clean_df,date_nunique,on=['order_id','order_id'])
clean_df=pd.merge(clean_df,hour_nunique,on=['order_id','order_id'])
clean_df=pd.merge(clean_df,status,on=['order_id','order_id'])
clean_df=pd.merge(clean_df,service,on=['order_id','order_id'])
clean_df=pd.merge(clean_df,pings,on=['order_id','order_id'])

clean_df=clean_df.drop('order_id',axis=1)


clean_df.head()
```

```python
clean_df[['altitude_max','altitude_min','altitude_avg']] = clean_df[['altitude_max','altitude_min','altitude_avg']].fillna(0)
clean_df[['longitude_max','longitude_min','longitude_avg','latitude_max','latitude_min','latitude_avg','seconds_max','seconds_min','seconds_avg']] = clean_df[['longitude_max','longitude_min','longitude_avg','latitude_max','latitude_min','latitude_avg','seconds_max','seconds_min','seconds_avg']].fillna(value=clean_df[['longitude_max','longitude_min','longitude_avg','latitude_max','latitude_min','latitude_avg','seconds_max','seconds_min','seconds_avg']].mean())
```
