```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('kaggle_Interests_group.csv')
pd.options.display.max_columns = None
```

```python
for col in df:
    if df[col].nunique()>1:
        print(col)
    else:
        pass
```

```python
a=[]
for col in df:
    if df[col].nunique()==1:
        a.append(col)
    else:
        pass
df[a] = df[a].fillna(0)


```

```python
df.isnull().sum()
```

```python
threshold=0.1*6340

b=[]
for col in df:
    if df[col].isnull().sum()>threshold:
        b.append(col)
    else:
        pass
df=df.drop(b,axis=1)
```

```python
c=[]
for col in df:
    if (df[col].isnull().sum()<threshold) & (df[col].isnull().sum()>0):
        c.append(col)
    else:
        pass
df = df.dropna(subset=c)
```

```python
df
```

```python
for col in df:
    print(df[col].unique())
```

```python
df['group'] = df['group'].map({'C':0,'P':1,'R':2,'I':3})
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df.corr(),annot=True,cmap='viridis')
```

```python
plt.figure(figsize=(12,7))
df.corr()['group'].sort_values().plot(kind='bar')
```

```python
x=df.drop('group',axis=1).values 
y=df['group'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
scores=[]
for i in range(1,50):
  tree=DecisionTreeClassifier(max_depth = i) 
  tree.fit(x_train, y_train) 
  scores.append(tree.score(x_test,y_test)) 
plt.plot(range(1,50),scores) 
plt.show()
```

```python
tree=DecisionTreeClassifier(max_depth =5) 
tree.fit(x_train, y_train) 
tree.score(x_test,y_test)
```

```python
predictions = tree.predict(x_test) 
from sklearn.metrics import classification_report,confusion_matrix 
print(classification_report(y_test,predictions)) 
print(confusion_matrix(y_test,predictions))
```
