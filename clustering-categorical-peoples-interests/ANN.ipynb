```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('kaggle_Interests_group.csv')
```

```python
#see the distribution data per label (Oh no! the data with label R is skew)
df.info()
sns.countplot(x='group',data=df)
```

```python
df
```

```python
df.nunique().sum
```

```python
#see the distinct values per column
for col in df:
    print(df[col].unique())
```

```python
#see the columns that real missing value, since the dataset also use NaN to indicate the value equal to zero
for col in df:
    if df[col].nunique()>1:
        print(col)
    else:
        pass
```

```python
#fill the NaN value with zero, only for columns use NaN to indicate the value equal to zero
a=[]
for col in df:
    if df[col].nunique()==1:
        a.append(col)
    else:
        pass
df[a] = df[a].fillna(0)


```

```python
df.isnull().sum()
```

```python
#drop the columns with the number of missing value >10% of total data
threshold=0.1*6340

b=[]
for col in df:
    if df[col].isnull().sum()>threshold:
        b.append(col)
    else:
        pass
df=df.drop(b,axis=1)
```

```python
#drop the rows with the number of missing value from columns with the number of missing value <10% of total data
c=[]
for col in df:
    if (df[col].isnull().sum()<threshold) & (df[col].isnull().sum()>0):
        c.append(col)
    else:
        pass
df = df.dropna(subset=c)
```

```python
df
```

```python
for col in df:
    print(df[col].unique())
```

```python
df['group'] = df['group'].map({'C':0,'P':1,'R':2,'I':3})
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df.corr(),annot=True,cmap='viridis')
```

```python
plt.figure(figsize=(12,7))
df.corr()['group'].sort_values().plot(kind='bar')
```

```python
x=df.drop('group',axis=1).values 
y=df['group'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.utils import to_categorical
y_cat_train = to_categorical(y_train) 
y_cat_test = to_categorical(y_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping
```

```python
from tensorflow.keras.layers import Dropout
model = Sequential()
model.add(Dense(units=160,activation='relu',kernel_regularizer=tf.keras.regularizers.l1_l2(0.01)))

model.add(Dense(units=4,activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=10)

model.fit(x_train,y_cat_train,epochs=500,validation_data=(x_test,y_cat_test),verbose=1)
```

```python
losses = pd.DataFrame(model.history.history)
losses[['accuracy','val_accuracy']].plot()
losses[['loss','val_loss']].plot()
```

```python
print(model.metrics_names)
print(model.evaluate(x_test,y_cat_test,verbose=0))
```

```python
from sklearn.metrics import classification_report,confusion_matrix

predictions = model.predict_classes(x_test)
print(classification_report(y_test,predictions))
```

```python
confusion_matrix(y_test,predictions)
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(confusion_matrix(y_test,predictions),annot=True)
```
